# Checking for Imbalance

Now that we have cleaned up the data, one issue that we still face is the issue of imbalance. As you can see below, we have significantly more samples of ham sms messages compared to spam messages. Potentially, this could impact the accuracy of our prediction model.

I am curious to see the impacts of this imbalance on my model's accuracy. Therefore, I will be using two datasets to create my model. An untouched cleaned training dataset that we cleaned in the last notebook, and another where I will be using SMOTE to level the imbalance (which I will be doing in this notebook). Afterwords, I will create a model based off both of these training datasets to try to quantify the difference imbalance correction has.

### Contents:
* Visualizing Imbalance
* SMOTE 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
import string

import nltk


sns.set(style="darkgrid")
pd.options.mode.chained_assignment = None
```

```python
cleaned = pd.read_csv('cleaned_training.csv')
cleaned_labels = pd.read_csv('train_labels.csv')
cleaned.head()
```

## Visualizing Imbalance

```python
sns.countplot(x="class", data=cleaned)
```

```python
# Counts of Spam vs Ham to check for imbalance
counts = cleaned.groupby('class').count()[['sms']]
counts
```

```python
(counts.iloc[1]/counts.sms.sum())[0]
```

It is clear that there is an imbalance of values. This makes sense since spam emails should be as prevalent as normal emails. As a result, let's see if we can balance the training and see if this will impact the accuracy of our model with the testing data.



## SMOTE

When trying to find ways to combat the issue of imbalance, I came across the SMOTE Algorithm. Therefore, I will be using SMOTE to combat imbalance within this dataset and see how it impacts my accuracy of different models

Link: https://imbalanced-learn.readthedocs.io/en/stable/generated/imblearn.over_sampling.SMOTE.html

```python
from imblearn.over_sampling import SMOTE

# Getting the Labels and corresponding data from cleaned
x_vals = cleaned.drop(columns = {'sms', 'class'})
y_vals = cleaned_labels
y_vals

# Setting SMOTE to random_state = 40 for reciprocation
smote = SMOTE(random_state=40)

# Using Smote to Generate synthetic datapoints
x_res, y_res = smote.fit_resample(x_vals, y_vals)

# Seeing new count of rows
x_res.shape
```

```python
x_res.head()
```

```python
y_res.head()
```

```python
# After SMOTE, there is no longer an imbalance in sms samples
temp = x_res.join(y_res)
temp.groupby('spam').count()[['char_length']].rename( columns = {'char_length' : 'count'})
```

```python
# Exporting our balanced dataset
x_res.to_csv('balanced_training.csv', index = False)
y_res.to_csv('balanced_labels.csv', index = False)
```
